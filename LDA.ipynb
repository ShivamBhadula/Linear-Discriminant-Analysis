```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Iris.csv")
dataset
```

```python
dataset['Species'].unique()
```

```python
for  ind in dataset['Species'].index:
    if dataset['Species'][ind]=='Iris-setosa':
        dataset['Species'][ind]=1
    elif dataset['Species'][ind]=='Iris-versicolor':
        dataset['Species'][ind]=2
    elif dataset['Species'][ind]=='Iris-virginica':
        dataset['Species'][ind]=3
    
```

```python
dataset
```

```python
X=dataset.iloc[:,:-1]
X
```

```python
y=dataset.iloc[:,-1].astype(int)
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
```

```python
X_test=sc.transform(X_test)
X_train
```

```python
X_test
```

```python
#apply LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda=LDA(n_components=2)
X_train=lda.fit_transform(X_train,y_train)
X_test=lda.transform(X_test)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
#Visualize training set results
from matplotlib.colors import ListedColormap
X_set,y_set=X_train,y_train
X1,X2=np.meshgrid(np.arange(start=X_set[:,0].min()-1,stop=X_set[:,0].max()+1,step=0.25),
np.arange(start=X_set[:,1].min()-1,stop=X_set[:,1].max()+1,step=0.25),)
plt.contourf(X1,X2,lr.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape),alpha=0.75,cmap=ListedColormap(('red','blue','green')))
plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())

for i,j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set==j,0],X_set[y_set==j,1],
    c=ListedColormap(('red','blue','green'))(i),label=j)


plt.title('Training set visualization')
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.legend()
plt.show()
```

```python
#Visualize test set results
from matplotlib.colors import ListedColormap
X_set,y_set=X_test,y_test
X1,X2=np.meshgrid(np.arange(start=X_set[:,0].min()-1,stop=X_set[:,0].max()+1,step=0.25),
np.arange(start=X_set[:,1].min()-1,stop=X_set[:,1].max()+1,step=0.25),)
plt.contourf(X1,X2,lr.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape),alpha=0.75,cmap=ListedColormap(('red','blue','green')))
plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())

for i,j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set==j,0],X_set[y_set==j,1],
    c=ListedColormap(('red','blue','green'))(i),label=j)


plt.title('Test set visualization')
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.legend()
plt.show()
```
